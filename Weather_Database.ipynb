```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a set of random latitude and longitude combos
lats = np.random.uniform(low=-90.0, high=90.0, size=1500)
lngs = np.random.uniform(low=-180.0, high=180.0, size=1500)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Add the lats and longs to a list
coordinates = list(lat_lngs)
```

```python
# Use the citipy module to determine city based on lat and long
from citipy import citipy
```

```python
# Create a list for holding the cities.
cities = []
# Identify the nearest city for each lat and long combo
for cooridnate in coordinates:
    city = citipy.nearest_city(cooridnate[0], cooridnate[1]).city_name
    
    # If the city is unique, then we will add it to the cities list
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count
len(cities)
```

```python
from datetime import datetime
```

```python
# Import requests library
import requests

# Import the API key
from config import weather_api_key
```

```python
# Starting URL for weather map API call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
# Create an empty list to hold the weather data
weather_city_data = []
# Print the beginning of the logging
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters
record_count = 1
set_count = 1

# Loop through all the cities in the list
for i, city in enumerate(cities):
    
    # Group cities in sets of 50 for logging purposes
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    # Create endpoint URL with each city.
    weather_city_url = url + "&q=" + city
    
    # Log URL, record, and set numbers and the city
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    # Add 1 to the record
    record_count += 1
    
    # Run and API request for each of the cities
    try:
        # Parse the JSON and retrieve data
        city_weather_module = requests.get(weather_city_url).json()
        #Parse out the needed data.
        city_lat = city_weather_module["coord"]["lat"]
        city_lng = city_weather_module["coord"]["lon"]
        city_max_temp = city_weather_module["main"]["temp_max"]
        city_humidity = city_weather_module["main"]["humidity"]
        city_clouds = city_weather_module["clouds"]["all"]
        city_wind = city_weather_module["wind"]["speed"]
        city_description = city_weather_module["weather"][0]["description"]
        try:        
            city_rain = city_weather_module["rain"]["3h"]
        except:
            city_rain = 0
        try:
            city_snow = city_weather_module["snow"]["3h"]
        except:
            city_snow = 0
        city_country = city_weather_module["sys"]["country"]
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather_module["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list
        weather_city_data.append({"City": city.title(), "Country": city_country, "Date": city_date, "Lat": city_lat, "Lng": city_lng, "Max Temp": city_max_temp, "Humidity": city_humidity, "Cloudiness": city_clouds, 
                          "Wind Speed": city_wind, "Current Description": city_description, "Rain inches (last 3 hrs)": city_rain, "Snow inches (last 3 hrs)": city_snow})

    # If an error is experienced, skip the city
    except:
        print("City not found. Skipping...")
        pass
    # If no rain
        #except:
        
        
# Indicate that the Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Convert array of dictionaries to a Pandas DF
weather_df = pd.DataFrame(weather_city_data)
weather_df.head()
```

```python
# Create the outputfile (CSV)
output_data_file = "challenge_data/WeatherPy_challenge.csv"
# Export the weather_df into a CSV
weather_df.to_csv(output_data_file, index_label="City_ID")
```
