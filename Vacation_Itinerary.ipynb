```python
# Import dependencies
import pandas as pd
import gmaps
import requests
import gmaps.datasets
# Import the API key
from config import g_key
# Configure gmaps to use your Google API key
gmaps.configure(api_key=g_key)
```

```python
# Store CSV from part 2 into a DataFrame
hotel_df = pd.read_csv('challenge_data/WeatherPy_vacation.csv')
hotel_df
```

```python
# Get hotels in close proximity 
hotel_df_one = hotel_df.loc[(hotel_df["Country"] == "TH")]
hotel_df_two = hotel_df.loc[(hotel_df["Country"] == "MM")]
hotel_df_three = hotel_df.loc[(hotel_df["Country"] == "IN")]
hotel_df_four = hotel_df.loc[(hotel_df["Country"] == "BD")]
```

```python
new_hotel_df = hotel_df_one[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng", "Hotel Name"]].copy()
new_two = new_hotel_df.append(hotel_df_two[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng", "Hotel Name"]].copy())
new_three = new_two.append(hotel_df_three[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng", "Hotel Name"]].copy())
final_hotel_df = new_three.append(hotel_df_four[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng", "Hotel Name"]].copy())
final_hotel_df
```

```python
# Get lat,lng for each hotel
rayong = (hotel_df_one["Lat"].values[0], hotel_df_one["Lng"].values[0])
falam = (hotel_df_two["Lat"].values[0], hotel_df_two["Lng"].values[0])
thenzawl = (hotel_df_three["Lat"].values[1], hotel_df_three["Lng"].values[1])
lalmohan = (hotel_df_four["Lat"].values[0], hotel_df_four["Lng"].values[0])
```

```python
# Create template for info box
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in final_hotel_df.iterrows()]
```

```python
# Create route between each hotel and make it a gmaps layer
rayong_lalmohan = gmaps.directions_layer(rayong, lalmohan, waypoints=[falam, thenzawl], travel_mode='DRIVING')
```

```python
# Congifure gmaps
fig = gmaps.figure(center=(22,93), zoom_level=4)
fig.add_layer(rayong_lalmohan)
fig
```

```python
locations = final_hotel_df[["Lat","Lng"]]
fig = gmaps.figure(center=(22,93), zoom_level=4)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
fig
```
